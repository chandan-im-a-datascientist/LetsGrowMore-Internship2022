**Dropping un required columns**

```python
dt_data= dt_data.drop('Id', axis=1)
```

```python
dt_data
```

**Data Preprocessing**

```python
from sklearn.preprocessing import LabelEncoder    #used to standardize the categorial value
le=LabelEncoder()
```

```python
dt_data['Species']=le.fit_transform(dt_data['Species'])   #changing categorial values into numbers
```

```python
dt_data
```

**Spliting the data**

```python
x=dt_data.iloc[:,:-1]
y=dt_data.iloc[:,-1]
x.head()
```

```python
y.head()
```

****Train and Test split****

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test= train_test_split(x,y,test_size=0.20,random_state=50)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
x_train.shape
```

```python
y_test.shape
```

**DECISION TREE ALGORITHM**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_predict=dt.predict(x_test)
```

```python
y_test=np.array(y_test)
y_test
```

**Accuracy**

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
acc=accuracy_score(y_predict,y_test)*100
acc
```

```python
confusion_matrix(y_predict,y_test)
```

# **Classificion Report**

```python
from sklearn.metrics import classification_report
print(classification_report(y_predict,y_test))
```
