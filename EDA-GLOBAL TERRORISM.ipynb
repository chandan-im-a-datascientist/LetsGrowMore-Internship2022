**Cleaning DATA**

```python
gt_data= gt_data[["iyear","imonth","iday","country_txt","region_txt","provstate","city","latitude","longitude","location","summary","attacktype1_txt","targtype1_txt","gname",
                 "motive","weaptype1_txt","nkill","nwound","addnotes"]]
gt_data.head()
```

```python
gt_data.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country","region_txt":"Region","provstate":"State","city":"City","latitude":"Latitude",
                        "longitude":"Longitude","location":"Location","summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type","gname":"Group Name",
                 "motive":"Motive","weaptype1_txt":"Weapon Type","nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
```

```python
gt_data.head()
```

```python
gt_data.isnull().sum()
```

```python
gt_data["Killed"]=gt_data["Killed"].fillna(0)
gt_data["Wounded"]=gt_data["Wounded"].fillna(0)
gt_data["Casuality"]=gt_data["Killed"]+gt_data["Wounded"]
```

```python
gt_data.describe()
```

# **EDA-VISUALIZATION**

**Number of attack per Year**

```python
attacks=gt_data["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

```python
attacks.plot(kind="bar",color="Orange",figsize=(15,15),fontsize=17)
plt.title("Year Wise Attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

```python
killed=gt_data[["Year","Killed"]].groupby("Year").sum()
killed.head()
```

**Graph**

```python
killed.plot(kind="bar",color="blue",figsize=(30,15))
plt.title("People Killed in each Year")
plt.xlabel("Years")
plt.ylabel("Number of People Killed")
```

```python
Wounded=gt_data[["Year","Wounded"]].groupby("Year").sum()
gt_data.head()
```

```python
Wounded.plot(kind="bar",color="skyblue",figsize=(30,15))
plt.title("People Wounded in each Year")
plt.xlabel("Years")
plt.ylabel("Number of People Wounded")
```

****Total attacks from 1970-2017****

```python
reg=pd.crosstab(gt_data.Year,gt_data.Region)
reg.head()
```

```python
reg.plot(kind="area", stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,10))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

**ATTACKS COUNTRY WISE**

**Group Wise Attacks**
